```python
files = r"C:\Users\imoha\OneDrive\Documents\MA\data\test_data\DoubleLaneChange"
names = r"C:\Users\imoha\OneDrive\Documents\MA\data\test_data\DoubleLaneChange\names_of_files_DoubleLaneChange.txt"
```

```python
# Iterate over the files in the directory
import os

for filename in os.listdir(files):
    if os.path.isfile(os.path.join(files, filename)):
        # Check if "80" is in the filename
        if '80' in filename:
            # Remove "80" from the filename
            new_filename = filename.replace('80', '')
            
            # Get the full paths of the old and new filenames
            old_path = os.path.join(files, filename)
            new_path = os.path.join(files, new_filename)
            
            # Rename the file
            os.rename(old_path, new_path)
            print(f'Renamed: {filename} to {new_filename}')
```
